```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("car_sales_data.xlsx")
data
```

```python
data = pd.DataFrame(data)
data
```

```python
male = data["Gender"].value_counts()
male
```

```python
male.head(10).plot(kind='bar')
plt.title('Gender')
plt.xlabel('Male')
plt.ylabel('Female')
plt.savefig('Gender', dpi=300, transparent=False)
plt.show()
```

```python
income_gender = data.groupby("Gender")["Annual Income"].sum()
income_gender

```

```python
income_gender.plot(kind="bar")
plt.title('Gender')
plt.xlabel("Female")
plt.ylabel("Male")
plt.show()
```
