```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
data = pd.read_excel("car_sales_data.xlsx")
data
```

```python
data = pd.DataFrame(data)
```

```python
table = pd.pivot_table(data=data,index='Company',values='Price ($)', aggfunc=np.sum)
table.head(10)
```

```python
table.head(10).plot(kind='bar')
plt.title('Total Price by Company')
plt.xlabel('Company')
plt.ylabel('Price($)')
plt.savefig('Total Price by Company', dpi=300, transparent=False)
plt.show()
```
