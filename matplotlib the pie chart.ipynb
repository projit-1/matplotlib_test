```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("car_sales_data.xlsx")
data
```

```python
data = pd.DataFrame(data)
```

```python
table = pd.pivot_table(data=data,index='Company',values='Price ($)', aggfunc=np.sum)
table.head(10)
```

```python
table.head(10).plot(kind='pie', y='Price ($)', autopct='%1.1f%%', figsize=(6, 6), legend=False)
plt.title('Total Price Distribution by Company')
plt.ylabel('')
plt.savefig('Total Price Distribution by Company', dpi=300, transparent=False)
plt.show()
```
